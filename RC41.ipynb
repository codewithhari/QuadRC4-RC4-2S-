```python
import time
import random
k = [i for i in range(2000)]
l = len(k)
random.shuffle(k)

iv1 = [i for i in range(2000)]
iv2 = [i for i in range(2000)]
iv3 = [i for i in range(2000)]
iv4 = [i for i in range(2000)]

N = 256
S1 = [i for i in range(0,N//3)]
S2 = [i for i in range(N//3,2*N//3)]
S3 = [i for i in range(2*N//3,N)]

k1 = k[::]
random.shuffle(iv1)
k2 = k[::-1]
random.shuffle(iv2)
k3 = k[0:l//2:-1] + k[l//2::-1]
random.shuffle(iv3)
k4 = k3[::-1]
random.shuffle(iv4)




def create_stateTable(s,k,iv):
    st = s[::]
    n = len(st)
    j = 0
    l = len(k)
    for i in range(n):
        j = (j+st[i]+k[i%l])%n
        st[i],st[j] = st[j],st[i]

    for i in range(n//2-1,-1,-1):
        j = ((j+st[i])^(k[i%l]+iv[i%l])) % n
        st[i],st[j] = st[j],st[i]
    for i in range(n//2,n):
        j = ((j+st[i])^(k[i%l]+iv[i%l])) % n
        st[i],st[j] = st[j],st[i]


    for y in range(n):
        if y%2 == 0:
            i = y//2
        else:
            i = (n - (y+1)//2)
        j = (j+st[i]+k[i%l])%n
        st[i],st[j] = st[j],st[i]
    
    return st

   
s1 = create_stateTable(S1,k1,iv1)
s2 = create_stateTable(S2,k1,iv1)
s3 = create_stateTable(S3,k1,iv1)
s4 = create_stateTable(S1,k2,iv2)
s5 = create_stateTable(S2,k2,iv2)
s6 = create_stateTable(S3,k2,iv2)
s7 = create_stateTable(S1,k3,iv3)
s8 = create_stateTable(S2,k3,iv3)
s9 = create_stateTable(S3,k3,iv3)
s10 = create_stateTable(S1,k4,iv4)
s11 = create_stateTable(S2,k4,iv4)
s12 = create_stateTable(S3,k4,iv4)
```

```python
op = 105200 # no of output bytes
i = 0
j = [0 for i in range(13)]
n = min(len(s1),len(s2),len(s3))
ans = ''
start_time = time.time()
for _ in range(op):
    i = (i+1)%n
    j[1] = (j[1]+s1[i]) % n
    s1[i],s2[j[1]] = s2[j[1]],s1[i]
    t1 = s1[(s1[i]+s1[j[1]])%n]
    
    j[2] = (j[2]+s2[i]) % n
    s2[i],s3[j[2]] = s3[j[2]],s2[i]
    t2 = s2[(s2[i]+s2[j[2]])%n]
    
    j[3] = (j[3]+s3[i]) % n
    s3[i],s4[j[3]] = s4[j[3]],s3[i]
    t3 = s3[(s3[i]+s3[j[3]])%n]
    
    j[4] = (j[4]+s4[i]) % n
    s4[i],s5[j[4]] = s5[j[4]],s4[i]
    t4 = s4[(s4[i]+s4[j[4]])%n]
    
    j[5] = (j[5]+s5[i]) % n
    s5[i],s6[j[5]] = s6[j[5]],s5[i]
    t5 = s5[(s5[i]+s5[j[5]])%n]
    
    j[6] = (j[6]+s6[i]) % n
    s6[i],s7[j[6]] = s7[j[6]],s6[i]
    t6 = s6[(s6[i]+s6[j[6]])%n]
    
    j[7] = (j[7]+s7[i]) % n
    s7[i],s8[j[7]] = s8[j[7]],s7[i]
    t7 = s7[(s7[i]+s7[j[7]])%n]
    
    j[8] = (j[8]+s8[i]) % n
    s8[i],s9[j[8]] = s9[j[8]],s8[i]
    t8 = s8[(s8[i]+s8[j[8]])%n]
    
    j[9] = (j[9]+s9[i]) % n
    s9[i],s10[j[9]] = s10[j[9]],s9[i]
    t9 = s9[(s9[i]+s9[j[9]])%n]
    
    j[10] = (j[10]+s10[i]) % n
    s10[i],s11[j[10]] = s11[j[10]],s10[i]
    t10 = s10[(s10[i]+s10[j[10]])%n]
    
    j[11] = (j[11]+s11[i]) % n
    s11[i],s12[j[11]] = s12[j[11]],s11[i]
    t11 = s11[(s11[i]+s11[j[11]])%n]
    
    j[12] = (j[12]+s12[i]) % n
    s12[i],s1[j[12]] = s1[j[12]],s12[i]
    t12 = s12[(s12[i]+s12[j[12]])%n]
    
#     print(t1,t2,t3,t4,t5,t6,t7,t8,t9,t10,t11,t12)
    ans += f'{t1:08b}'+f'{t2:08b}'+f'{t3:08b}'+f'{t4:08b}'+f'{t5:08b}'+f'{t6:08b}'+f'{t7:08b}'+f'{t8:08b}'+f'{t9:08b}'+f'{t10:08b}'+f'{t11:08b}'+f'{t12:08b}'
#     ans += f'{t1:08b}'+f'{t2:08b}'+f'{t3:08b}'

print("--- %s seconds ---" % (time.time() - start_time))

print(len(ans))
output = open('output.txt', 'w')
output.write(ans[96001:10096001])
output.close()
# print(ans)
    
    
```
